```python
#there is no description for parameters on kaggle so I'm deleting all parameters that are not intuitive to use (e.g. how it is calculated the location)
X = data.drop(['Type', 'Location', 'Character','Conscience', 'Duration', 'Intensity', 'Visual', 'Sensory', 'DPF', 'Defect'], axis = 1) 
y = data['Type']
```

```python
encoding = LabelEncoder() #we cannot use the string for our classifier so I'm encoding it
y = encoding.fit_transform(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 0)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
y_pred = model.predict(X_test) 
```

```python
y_true = y_test
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average= 'weighted')
print("Accuracy:", accuracy)
print("Precision:", precision)
```
